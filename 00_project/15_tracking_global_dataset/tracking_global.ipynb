<h2 style="font-family:Comic Sans MS; color:purple">2. Import Libraries</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2 style="font-family:Comic Sans MS; color:purple">3. Load the Dataset</h2>

```python
df = pd.read_csv('./data/electric-car-sales-share NEW.csv')
```

```python
df.head()
```

<h2 style="font-family:Comic Sans MS; color:purple">4. Data Analysis</h2>

```python
df.info()
```

<h2 style="font-family:Comic Sans MS; color:purple">5. Ploting</h2>

```python
# Bar Plot: Average share of new cars that are electric for each year
plt.figure(figsize=(40, 6))
plt.subplot(1, 3, 1)
sns.barplot(data=df, x='Year', y='Share of new cars that are electric', palette='muted')
plt.title('Average Share of New Cars that are Electric')
plt.xlabel('Year')
plt.ylabel('Average Share')
plt.grid(True)
```

```python
# Box Plot: Distribution of the share of new cars that are electric for each year
plt.figure(figsize=(40, 6))
plt.subplot(1, 3, 2)
sns.boxplot(data=df, x='Year', y='Share of new cars that are electric', palette='pastel')
plt.title('Distribution of Share of New Cars that are Electric')
plt.xlabel('Year')
plt.ylabel('Share')
plt.grid(True)
```

```python
# Histogram: Distribution of the share of new cars that are electric
plt.figure(figsize=(40, 6))
plt.subplot(1, 3, 1)
sns.histplot(data=df, x='Share of new cars that are electric', bins=10, kde=True, color='skyblue')
plt.title('Distribution of Share of New Cars that are Electric')
plt.xlabel('Share of New Cars')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Pairplot: Pairwise relationships and histograms
sns.pairplot(df.drop(['Entity', 'Code'], axis=1))
plt.suptitle('Pairwise Relationships and Histograms', y=1.02)
plt.show()
```

```python
# Jointplot: Relationship between 'Year' and 'Share of new cars that are electric'
sns.jointplot(data=df, x='Year', y='Share of new cars that are electric', kind='scatter')
plt.suptitle('Jointplot: Year vs Share of new cars that are electric', y=1.02)
plt.show()
```
