```python
import pandas as pd 
```

```python
data = {
    'date': ['2021-12-01', '01-12-2022', '2022/12/01', '12-01-2021'],
    'country': ['USA', 'U.S.A.', 'America', 'United States'],
    'name': ['Sibtain', 'Subtain', 'Tayyab', 'Tyyab'],
    'sales_2020': [100, 200, None, 200],
    'sales_2021': [None, 150, 300, 150]
}
# make pandas dataframe
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# standardizing the data format
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

# Assignment: Impute fill the date and try another method to complete the date formate inconsistencies

```python
# Impute Missing Dates using forward filling
df['date'] = pd.to_datetime(df['date'], errors='coerce')  # Convert 'date' column to datetime
df['date'] = df['date'].fillna(method='ffill')

# Standardize Date Formats
df['date'] = df['date'].dt.strftime('%Y-%m-%d')  # Convert 'date' column to desired format

# Validate and Verify
print(df['date'].head(10))  # Print the first 10 dates to verify the results
df.head()
```

```python
# Harmonize the name of the country 
country_mapping = {'USA': 'United States', 'U.S.A.': 'United States', 'America': 'United states'}
df['country'] = df['country'].replace(country_mapping)
df.head()
```

```python
# Correct the typographical Mistakes in name 
# let's assume we want to correct 'Subtain': 'Sibtain', 'Tyyab': 'Tayyab'
df['name'] = df['name'].replace({'Subtain': 'Sibtain', 'Tyyab': 'Tayyab'})
df.head()
```

```python
df = df.drop_duplicates(subset='name')
df.head()
```
