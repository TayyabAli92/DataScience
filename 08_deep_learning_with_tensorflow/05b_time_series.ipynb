# Time Series Data Analysis Project

```python
# import libraries
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# read the data
df = pd.read_csv('AirPassengers.csv')
df
```

# Data Pre-processing

```python
# changes the names of columns
df.columns = ["Date", "Number of Passengers"]
df.head()
```

```python
sns.lineplot(x = "Date", y = "Number of Passengers",  data = df)
```

```python
# plot the data using sns
sns.set(rc={'figure.figsize':(15, 5)})
sns.lineplot(x = "Date", y = "Number of Passengers",  data = df)
# set the labels
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.title("Number of Passengers Over Time")
# rotate x-axis ticks
plt.xticks(rotation=90, fontsize=6)

# make same plot using plotly
import plotly.express as px
fig = px.line(df, x="Date", y="Number of Passengers", title='Number of Passengers Over Time')
# change color of line
fig.update_traces(line_color='blue')
fig.show()
```

```python
# make two sided plot using matplolib
fig, ax = plt.subplots(figsize=[15, 5])
ax.plot(df["Date"], df["Number of Passengers"], color='blue')
ax.set_xlabel("Date", fontsize=14)
ax.set_ylabel("Number of Passengers", color='blue', fontsize=14)
ax.tick_params(axis="y", color='blue', labelsize=14)
ax.tick_params(axis="x", labelsize=6, rotation=90)
ax2 = ax.twinx()
ax2.plot(df["Date"], df["Number of Passengers"], color='red')
ax2.set_ylabel('Number of Passengers', color='red', fontsize=14)
ax2.tick_params(axis='y', colors='red', labelsize=14)
plt.show()
```

```python
x = df['Date'].values
y1 = df['Number of Passengers'].values

# Plot
fig, ax = plt.subplots(1, 1, figsize=(16,5), dpi= 120)
plt.fill_between(x, y1=y1, y2=-y1, alpha=0.5, linewidth=2, color='seagreen')
plt.ylim(-800, 800)
plt.title('Air Passengers (Two Side View)', fontsize=16)
plt.hlines(y=0, xmin=np.min(df['Date']), xmax=np.max(df['Date']), linewidth=.5)
plt.show()
```

```python
# decompose the data
result = seasonal_decompose(df['Number of Passengers'], model='multiplicative', period=12)
result.plot()
# plot in plotly
fig = px.line(result.observed, x=result.observed.index, y=result.observed.values, title='Observed')
fig.show()
```
